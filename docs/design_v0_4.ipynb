# Design: Architecture and Components

This notebook explains the architecture of `weather-model-graphs` version 0.4.0 with the new extensions.

## Core Principles

1. **NetworkX-First**: Use `networkx.DiGraph` as primary data structure
2. **Modularity**: Each step outputs a graph object
3. **Multi-Backend Support**: Convert to/from PyTorch Geometric and DGL
4. **Production Ready**: Configuration, features, and ML pipelines built-in

## Quality Metrics by Architecture

Performance characteristics of different architectures:

```python
import weather_model_graphs as wmg
import time

grid_sizes = [16, 32, 64]
architectures = ['keisler', 'graphcast', 'meshgraphnet']

results = []
for arch in architectures:
    for grid_size in grid_sizes:
        try:
            start = time.time()
            graph = wmg.load_prebuilt(arch, grid_size=grid_size)
            elapsed = time.time() - start
            
            nodes = graph.number_of_nodes()
            edges = graph.number_of_edges()
            density = edges / (nodes ** 2) * 100 if nodes > 0 else 0
            
            results.append({
                'Architecture': arch,
                'Grid': f"{grid_size}x{grid_size}",
                'Nodes': nodes,
                'Edges': edges,
                'Density': f"{density:.2f}%",
                'Time (ms)': f"{elapsed*1000:.1f}",
            })
        except Exception as e:
            pass

df = pd.DataFrame(results)
print("\nPerformance Comparison:")
print(df.to_string(index=False))
print("\nNote: Density = edges / (nodes²) × 100%")
print("Higher density = more connections (more expensive, more expressive)")
```
